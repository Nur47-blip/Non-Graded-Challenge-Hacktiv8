```python
Task

Create a library catalog system using OOP. The system should allow users to add books to the catalog, 
search for books by title or author, and remove books from the catalog.
    
Requirements:
1.Each book should have a title, author, and unique identifier
2.The system should provide the following functionality:
  ● Adding a book to the catalog.
  ● Searching for books by title or author.
  ● Removing a book from the catalog.
  ● Displaying all books in the catalog.
3.Do the exception handling
```

```python
class Catalog:
    def __init__(self):
        self.books = []

    def add_book(self, id, title, author):
        # Jika sudah terdapat buku dengan id yang sama, raise error
        try:
            for book in self.books: 
                if book['book_id'] == id:
                    raise ValueError()
        except:
            print("Book ID has been used.")
        else:
            self.books.append({'book_id':id, 'book_title':title, 'book_author':author})
            print("Book successfully added.")

    def search_by_title(self, title):
        # Jika buku tidak ditemukan, raise error
        result = None
        try:
            for book in self.books: 
                if book['book_title'] == title:
                    result = book
                    break
            if result == None:
                raise ValueError()
        except:
            print(f"Book with the title `{title}` is not found.")
        else:
            print(result)

    
    def search_by_author(self, author):
        # Jika buku tidak ditemukan, raise error
        result = None
        try:
            for book in self.books:
                if book['book_author'] == author:
                    print(book)
                    return
            if result == None:
                raise ValueError()
        except:
            print(f"Book with the author `{author}` is not found.")
        else:
            print(result)

    def remove(self, id):
        # Jika buku tidak ditemukan, raise error
        result = None
        try:
            for book in self.books:
                if book["book_id"] == id:
                    result = book
                    break
            if result == None:
                raise ValueError()
        except:
            print(f"Book with ID {id} is not found.")
        else:
            self.books.remove(book)

    def display(self):
        # Jika self.books kosong, raise error
        try:
            if len(self.books) == 0:
                raise ValueError()
        except:
            print("Current Library is empty.")
        else:
            for book in self.books:
                print(book)

myLibrary = Catalog()
myLibrary.display()
myLibrary.add_book(1, "Doraemon", "Fujiko F. Fujio")
myLibrary.add_book(2, "Naruto", "Masashi Kishimoto")
myLibrary.add_book(3, "Ranah 3 Warna", "Ahmad Fuadi")
myLibrary.add_book(3, "One Piece", "Eiichiro Oda")
myLibrary.add_book(4, "The Art of Computer Programming", "Donald Knuth")
myLibrary.search_by_title("Naruto")
myLibrary.search_by_title("Attack on Titan")
myLibrary.search_by_author("Masashi Kishimoto")
myLibrary.search_by_author("Hajime Isayama")
myLibrary.remove(1)
myLibrary.remove(10)
myLibrary.display()
    
```
