## Data Loading

Mengambil data dari BigQuery dan memasukkannya ke dalam dataframe.

```python
from google.colab import auth
from google.cloud import bigquery
auth.authenticate_user()
print('Authenticated')

project_id = "playground-397709" #GUNAKAN GCP PROJECT-ID KALIAN MASING-MASING
client = bigquery.Client(project=project_id)
```

```python
df = client.query('''
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
''').to_dataframe()
df
```

## Exploration

Berikut adalah query untuk mendapatkan seluruh trip dengan `bike_id` tidak sama dengan null dan durasi trip melebihi 1000 menit.

```python
df = client.query('''
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE bike_id IS NOT NULL AND duration_minutes > 1000
''').to_dataframe()
df
```

Berikut ini adalah query untuk mendapatkan seluruh trip dengan `subscriber_type` adalah "Walk Up" dan `end_station_name` adalah "Stolen"

```python
df = client.query('''
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE subscriber_type = "Walk Up" AND end_station_name = "Stolen"
''').to_dataframe()
df
```
